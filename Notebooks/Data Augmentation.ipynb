```python
def produce_transformations(matrix, threeD=False):
    if threeD:
        axis = (1,2)
    else:
        axis = (0,1)
        
    matrices = []
    for times in range(4):
        mat = np.rot90(matrix, times, axis)
        flipped_mat = np.flip(mat, 1)
        matrices += [mat, flipped_mat]
    return matrices

mats = produce_transformations(actions[0])
for mat in mats:
    print (mat, end='\n\n')
    
```
